```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('avocado.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime (df['Date'])
```

```python
df.info()
```

```python
df.query('year == 2015')
```

```python
df.loc[df['type'] == 'conventional']
df.loc[df['region'] == 'Albany'].head()
```

```python
df.groupby(['Date', 'Total Bags', 'Small Bags']).count().head(5)
```

```python
df_melted = pd.melt(df, id_vars ='Date',
                    value_vars = ['Total Bags', 'Small Bags'],
                    var_name = 'Bags', value_name = 'Count')
```

```python
df_melted
```

```python
sns.relplot(data = df_melted, kind='line', x = 'Date', y = 'Count', hue = 'Bags')
```

```python
pd.qcut(df['Total Volume'], q = 4,
        labels = ['Poor', 'Modest','Good','Excellent'])
```

```python
df['Sales_Volume'] = pd.qcut(df['Total Volume'], q = 4,
                            labels = ['Poor', 'Modest', 'Good', 'Excellent'])
```

```python
df.head()
```

```python
sns.countplot(df, x = "year", hue = "Sales_Volume")
```

```python
df_melted = pd.melt(df, id_vars = 'Date',
                    value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'],
                    var_name = 'Bags', value_name = 'Count')
```

```python
df_melted
```

```python
sns.relplot(data = df_melted, kind = 'line', x = 'Date', y = 'Count', hue = 'Bags')
```
